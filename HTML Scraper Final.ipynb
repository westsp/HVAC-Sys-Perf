```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import os

lol = []
directory18 = r'IECC Data\2018'

for filename in os.listdir(directory18):
    if filename.endswith('htm'):
        filepath = os.path.join(directory18, filename)
       
        # Prettify HTML document using BeautifulSoup
        with open(filepath) as html_doc:
            soup = BeautifulSoup(html_doc)
        soup.prettify()
        
        # Extract Building Type and Climate Zone Info
        heading = soup.find('body')
        para_search = heading.find('p').find_next_sibling('p').find_next_sibling('p').find_next_sibling('p')
        building_type = para_search.find_next('b').get_text()
        met_station = para_search.find_next_sibling('p').find_next('b').get_text()

        if '489000' in met_station:
            cz = '0A'
        elif '411940' in met_station:
            cz = '0B'        
        elif '911820' in met_station:
            cz = '1A'
        elif '421820' in met_station:
            cz = '1B'
        elif '747880' in met_station:
            cz = '2A'
        elif '722745' in met_station:
            cz = '2B'
        elif '722190' in met_station:
            cz = '3A'
        elif '722700' in met_station:
            cz = '3B'
        elif '722904' in met_station:
            cz = '3C'
        elif '744860' in met_station:
            cz = '4A'
        elif '723650' in met_station:
            cz = '4B'
        elif '727930' in met_station:
            cz = '4C'
        elif '725280' in met_station:
            cz = '5A'
        elif '724695' in met_station:
            cz = '5B'
        elif '727885' in met_station:
            cz = '5C'
        elif '726440' in met_station:
            cz = '6A'
        elif '727750' in met_station:
            cz = '6B'
        elif '727470' in met_station:
            cz = '7'
        elif '702610' in met_station:
            cz = '8'
        else:
            print('CZ Name Error')
        print(cz)

        # Extract End Use Table Values
        key_search = soup.find('b', string='End Uses')
        table_search = key_search.find_next_sibling('table')

        row_search_heating = table_search.find_next('td', string='Heating')
        heating_el = row_search_heating.find_next_sibling('td').get_text(strip=True)
        heating_ng = row_search_heating.find_next_sibling('td').find_next_sibling('td').get_text(strip=True)
        np.float(heating_el)
        np.float(heating_ng)

        row_search_cooling = table_search.find_next('td', string='Cooling')
        cooling_el = row_search_cooling.find_next_sibling('td').get_text(strip=True)
        np.float(cooling_el)

        row_search_IntLight = table_search.find_next('td', string='Interior Lighting')
        IntLight_el = row_search_IntLight.find_next_sibling('td').get_text(strip=True)
        np.float(IntLight_el)

        row_search_ExtLight = table_search.find_next('td', string='Exterior Lighting')
        ExtLight_el = row_search_ExtLight.find_next_sibling('td').get_text(strip=True)
        np.float(ExtLight_el)

        row_search_IntEquip = table_search.find_next('td', string='Interior Equipment')
        IntEquip_el = row_search_IntEquip.find_next_sibling('td').get_text(strip=True)
        IntEquip_ng = row_search_IntEquip.find_next_sibling('td').find_next_sibling('td').get_text(strip=True)
        np.float(IntEquip_el)
        np.float(IntEquip_ng)

        row_search_ExtEquip = table_search.find_next('td', string='Exterior Equipment')
        ExtEquip_el = row_search_ExtEquip.find_next_sibling('td').get_text(strip=True)
        np.float(ExtEquip_el)

        row_search_fans = table_search.find_next('td', string='Fans')
        fans_el = row_search_fans.find_next_sibling('td').get_text(strip=True)
        np.float(fans_el)

        row_search_pumps = table_search.find_next('td', string='Pumps')
        pumps_el = row_search_pumps.find_next_sibling('td').get_text(strip=True)
        np.float(pumps_el)

        row_search_heatrej = table_search.find_next('td', string='Heat Rejection')
        heatrej_el = row_search_heatrej.find_next_sibling('td').get_text(strip=True)
        np.float(heatrej_el)

        row_search_humid = table_search.find_next('td', string='Humidification')
        humid_el = row_search_humid.find_next_sibling('td').get_text(strip=True)
        np.float(humid_el)

        row_search_heatrec = table_search.find_next('td', string='Heat Recovery')
        heatrec_el = row_search_heatrec.find_next_sibling('td').get_text(strip=True)
        np.float(heatrec_el)

        row_search_swh = table_search.find_next('td', string='Water Systems')
        swh_el = row_search_swh.find_next_sibling('td').get_text(strip=True)
        swh_ng = row_search_swh.find_next_sibling('td').find_next_sibling('td').get_text(strip=True)
        np.float(swh_el)
        np.float(swh_ng)

        row_search_refrig = table_search.find_next('td', string='Refrigeration')
        refrig_el = row_search_refrig.find_next_sibling('td').get_text(strip=True)
        np.float(refrig_el)

        row_search_total = table_search.find_next('td', string='Total End Uses')
        total_el = row_search_total.find_next_sibling('td').get_text(strip=True)
        total_ng = row_search_total.find_next_sibling('td').find_next_sibling('td').get_text(strip=True)
        np.float(total_el)
        np.float(total_ng)

        # Create new list row for each loop iteration
        row = [2018, building_type, cz, heating_el, heating_ng, cooling_el, IntLight_el, ExtLight_el, IntEquip_el, 
               IntEquip_ng, ExtEquip_el, fans_el, pumps_el, heatrej_el, humid_el, heatrec_el, swh_el, swh_ng, 
               refrig_el, total_el, total_ng]
        print(row)
        # Append new list row to list of lists for each loop iteration
        lol.append(row)
        
    else:
        continue
        
# Write list of lists to a dataframe and export as an Excel file
df = pd.DataFrame(lol)
df.to_excel('IECC 2018 Data Scraper Output.xlsx')
print(df.head())
```
